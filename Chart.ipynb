```python
class Chart:
    def __init__(self, line_num=0, part_num=0):
        self.line_num = line_num
        self.part_num = part_num

        self.line_list = self._line(line_num, part_num)
        self.title = random.choice([myutil.rand_string(10), ''])
        self.xlabel = random.choice([myutil.rand_string(10), ''])
        self.ylabel = random.choice([myutil.rand_string(10), ''])
        
        self.lnum = len(self.line_list)
        
        #superiority
        self.superiority  = self._superiority(self.line_list)
        #partition num
        self.l1p_num, self.l2p_num = self._p_num(self.line_list)
        #monotonic
        self.l1_mon, self.l2_mon = self._monotonic(self.line_list)
        #minmax point
        self.l1_min, self.l2_min = self._min(self.line_list)
        self.l1_max, self.l2_max = self._max(self.line_list)
        #partition start end
        self.l1_se, self.l2_se = self._SE(self.line_list)
        #growth type
        self.l1_g, self.l2_g = self._growth_type(self.line_list)
        
  
    def _superiority(self, l_list):
        ret = 0
        if len(l_list) != 2:
            return np.nan
        
        l1 = l_list[0]
        l2 = l_list[1]
        
        if min(l1.y) >= max(l2.y):
            ret = 1 #1st line
        elif min(l2.y) >= max(l1.y):
            ret = 2 #2nd line
        else:
            ret = 0 #none
            
        return ret
    
    def _p_num(self, l_list):
        l1p_num = len(l_list[0].p_list)
        l2p_num = np.nan

        if len(l_list) == 2:
            l2p_num = len(l_list[1].p_list)
            
        #print("_p_conf return : ", conf)
        return l1p_num, l2p_num


    def _monotonic(self, l_list):
        l1_mon = l_list[0]._monotonic()
        l2_mon = np.nan
        
        if self.lnum == 2:
            l2_mon = l_list[1]._monotonic()
            
        return l1_mon, l2_mon
    
    #(S,O)
    def _min(self, l_list):
        l1_min_so = [0, 0]
        l2_min_so = [np.nan, np.nan]
        
        l = l_list[0]
        o = round(min(l.y),4)
        idx = np.argmin(l.y)
        
        s = round(l.x[idx],4)

        l1_min_so[1] = o
        l1_min_so[0] = s
        
        if self.lnum == 2:
            l = l_list[1]
            o = round(min(l.y),4)
            idx = np.argmin(l.y)

            s = round(l.x[idx],4)

            l2_min_so[1] = o
            l2_min_so[0] = s
        
        #print("minmax_so  : ", minmax_so)
        return l1_min_so, l2_min_so
    
    def _max(self, l_list):
        l1_max_so = [0, 0]
        l2_max_so = [np.nan, np.nan]
        
        l = l_list[0]
        o = round(max(l.y),4)
        idx = np.argmax(l.y)
        
        s = round(l.x[idx],4)

        l1_max_so[1] = o
        l1_max_so[0] = s
        
        if self.lnum == 2:
            l = l_list[1]
            o = round(max(l.y),4)
            idx = np.argmax(l.y)

            s = round(l.x[idx],4)

            l2_max_so[1] = o
            l2_max_so[0] = s
        
        #print("minmax_so  : ", minmax_so)
        return l1_max_so, l2_max_so
    
    def _SE(self, l_list):
        
        n = np.nan
        l1_se = [[n,n], [n,n,n,n], [n,n,n,n,n,n]]
        l2_se = [[n,n], [n,n,n,n], [n,n,n,n,n,n]]
         
        l = l_list[0]
        pnum = l.partition_num
        
        temp = []
        for p in l.p_list:
            #print("p.x : ", p.x)
            #print("type(p.x) : ", type(p.x))
            temp = temp + list(np.round(p.x, 4))
            
        l1_se[pnum-1] = temp
        #print("temp : ", temp)
        
        
        l1_se = sum(l1_se, [])  
        #print("s1s2_ : ", s1s2_)
        
        if self.lnum == 2:
            l = l_list[1]
            pnum = l.partition_num

            temp = []
            for p in l.p_list:
                #print("p.x : ", p.x)
                #print("type(p.x) : ", type(p.x))
                temp = temp + list(np.round(p.x, 4))

            l2_se[pnum-1] = temp
            #print("temp : ", temp)


        l2_se = sum(l2_se, [])  
            #print("s1s2_ : ", s1s2_)
        
        return l1_se, l2_se

    def _growth_type(self, l_list):
        n = np.nan
        l1_g = [[n], [n,n], [n,n,n]]
        l2_g = [[n], [n,n], [n,n,n]]
        
        l = l_list[0]

        idx = len(l.p_list)-1
        
        #print("line relation list")
        #print(line.rel_list)
        
        for i, p in enumerate(l.p_list):
            #print("sign : ", p.sign())
            l1_g[idx][i] = l.g_list_num[i]
            if p.sign() == -1:
                l1_g[idx][i] = l1_g[idx][i] + 1
                
        l1_g = sum(l1_g, [])
        #print("rel : ", rel_)
        
        if self.lnum == 2:
            l = l_list[1]

            idx = len(l.p_list)-1

            #print("line relation list")
            #print(line.rel_list)

            for i, p in enumerate(l.p_list):
                #print("sign : ", p.sign())
                l2_g[idx][i] = l.g_list_num[i]
                if p.sign() == -1:
                    l2_g[idx][i] = l2_g[idx][i] + 1

        l2_g = sum(l2_g, [])
            #print("rel : ", rel_)
        
        return l1_g, l2_g
  
        

    def _line(self, line_num, part_num):
        if line_num == 0:
            self.line_num = random.randint(1, 2)
            line_num = self.line_num
            
        line_list = []
        for i in range(line_num):
            label = myutil.rand_string(10)
            linewidth = random.randint(1, 4)
            
            if part_num == 0:
                l = Line(random.randint(1, 3), label, linewidth)
            else:
                l = Line(part_num, label, linewidth)
                
            line_list.append(l)
            
        return line_list
    
    def draw(self, fmt_list=None):
        plt.title(self.title, loc=random.choice(['center','left','right']))
        plt.xlabel(self.xlabel)
        plt.ylabel(self.ylabel)

        if fmt_list == None:
            fmt_list = ['' for x in range(len(self.line_list))]

        for i, line in enumerate(self.line_list):
            line.draw(fmt_list[i])
            
    def show(self):
        plt.show()
        
```
